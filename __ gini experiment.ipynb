```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
nums = np.random.chisquare(2, 200)
plt.hist( nums, bins=20 );
```

```python
cumsum = np.sort(nums)
```

```python
plt.plot(
    range(cumsum.shape[0]),
    cumsum
);
plt.plot(
    range(cumsum.shape[0]),
    np.linspace(0,cumsum[-1],cumsum.shape[0]),
)
```

```python
def gini2(x):
    # (Warning: This is a concise implementation, but it is O(n**2)
    # in time and memory, where n = len(x).  *Don't* pass in huge
    # samples!)

    # Mean absolute difference
    mad = np.abs(np.subtract.outer(x, x)).mean()
    # Relative mean absolute difference
    rmad = mad/np.mean(x)
    # Gini coefficient
    g = 0.5 * rmad
    return g
```

```python
X = nums
```

```python
def gini(X):
    st = np.sort(X)
    st = np.cumsum(st)
    st /= st[-1]
    vs = np.linspace(0, st[-1], st.shape[0]) #diagonal
    return np.sum(vs - st) / np.sum(vs)
```

```python
plt.plot(st)
plt.plot(vs)
```

```python
vs1,vs2 = [],[]
for NN in range(100, 1000, 10):
    nums = np.random.chisquare(2, NN)
    vs1.append(gini(nums))
    vs2.append(gini2(nums))

plt.plot(vs1)
plt.plot(vs2)
```
